### 5. Un grupo de N estaciones comparten un canal Aloha puro de 56Kbps. Cada estación entrega un mensaje de 1000 bits cada 100 segundos. ¿Cuál es el máximo valor de N?

$$ Rb_{\text{real}} = Rb * \text{eficiencia} $$

$$ 56000 b/s * 0.18 = 10080 b/s $$

$$ Rb_{\text{estación}} = 1000b / 100s = 10b/s $$

$$ N = Rb_{\text{real}} / Rb_{\text{estación}} $$

$$ N = 10080 / 10 = 1008 \text{ estaciones} $$

### 11. ¿Qué significa tipo de cableado 10Base-T empleado en Redes Ethernet? ¿Qué cableado se emplea en 10Base2 y 10Base-F?

* **10Base-T:** Par trenzado (twisted pair). Las estaciones se conectan con par trenzado al hub que facilita el mantenimiento de los cables. Longitud máxima de cable de 100 metros.
* **10Base2:** Coaxil delgado. 200 metros por segmento, 30 nodos por segmento.
* **10Base-F:** Fibra óptica. 2000 metros por segmento, 1024 nodos por segmento.

### 19. Enunciar dos ventajas y desventajas de los 3 tipos de LANs vistos

#### 802.3 Ethernet
* **Ventajas**
    * Protocolo sencillo, cable pasivo, sin modems
    * Retardo con baja carga es casi cero
* **Desventajas**
    * Trama mínima de 64 bytes
    * No determinístico

#### 802.4 Token Bus
* **Ventajas**
    * Más determinístico que 802.3
    * Puede manejar tramas mínimas cortas
* **Desventajas**
    * Protocolo complicado
    * Retardo grande cuando la carga es baja (se debe esperar el token incluso aunque no haya tráfico)
    
#### 802.5 Token Ring
* **Ventajas**
    * Conexiones PaP, sencillas y digitales
    * El uso de centros de cableado permite detectar y eliminar fallas de cableado automáticamente
* **Desventajas**
    * Función de Monitor centralizada
    * Retardo a muy baja carga (por paso del token)

### 27. Con una tasa de transmisión de 5Mbps y una velocidad de propagación de 200m/&mu;s, ¿a cuántos metros de cable equivale el retardo de un bit de una interfase Token Ring?

$$ 200m/\mu s = 200000km/s $$

$$ L_t = Rb * T_t $$

$$ \dfrac{L_t}{Rb} = T_t $$

$$ \dfrac{1b}{5000000b/s} = 2 * 10^{-7}s $$

$$ 2 * 10^{-7}s * 2*10^9m/s = 40m $$

40 metros
