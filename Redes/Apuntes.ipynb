# Apuntes

## Capas Modelo OSI

- Capa Física
- Capa de Enlace de Datos
- Capa de Red
- Capa de Transporte
- Capa de Sesión
- Capa de Presentación
- Capa de Aplicación

### Capa Física

Se ocupa de tx de bits por el canal de comunicación.

Diseño de aspectos mecánico eléctrico, de procedimiento de interfase y el medio de tx física, que se encuentra bajo la capa física.

Voltajes del bit 1 o 0, duración del bit, tx bidireccional, establecimiento y desconexión de comunicaciones, cantidad de pines en conectores.

### Capa de Enlace de Datos

A partir de una línea de tx común, transformarla en una línea sin errores para la capa de red.

Emisor divide entrada de datos en tramas, las transmite secuencialmente y procesa las tramas de reconocimiento devueltas por el receptor.

Emplea mecanismos de regulación de tráfico para evitar que tranmisor muy rápido sature a receptor lento.

Subcapa MAC en redes de difusión tienen control de acceso al medio compartido

### Capa de Red

Se ocupa del control de operación de la subred.

Determina como encaminar los paquetes de origen a destino con tablas estáticas o dinámicas de ruteo.

Controla congestión y contabilidad de paquetes.

Resuelve problemas de interconexión de redes heterogéneas.

### Capa de Transporte

Acepta datos de la capa de sesión, los divide en unidades más pequeñas, los pasa a la capa de red y se asegura de que lleguen correctamente al otro extremo.

Normalmente crea una conexión de red distinta para cada conexión de transporte solicitada, aunque según caudal puede crear múltiples conexiones de red. O si conexión es costosa puede multiplexar varias conexiones de transporte a la misma conexión de red.

Determina tipo de servicio a dar a capa de sesión (Canal P a P o Difusión de mensajes a múltiples destinos).

Provee mecanismos de control de flujo.

### Capa de Sesión

Permite a distintos dispositivos crear sesiones entre ellos.
,
Coordina sesiones en uno o más procesos corriendo en distintos hosts.

Una sesión permite transporte de datos, pero con mejores servicios que capa de transporte, como acceso a sistema remoto de tiempo compartido o transferencia de archivos entre dos máquinas.

Control del diálogo. Manejo de testigo (ficha o token). Sincronización

### Capa de Presentación

Realiza funciones de búsqueda de soluciones generales. Síntaxis y semántica de la información transmitida.

Codificación y compresión de datos. Criptografía.

### Capa de Aplicación

Contiene protocolos que se necesitan con frecuencia para interfase de procesos de usuarios.

Transferencia de archivos, correo electrónico, carga remota de trabajos, otros recursos de uso general.

## Tranmisión de datos en modelo OSI

**La tranmisión de datos es vertical, sin embargo cada capa se programa como sistema horizontal.**
