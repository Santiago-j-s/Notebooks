```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1, 20)
y = x ** 2

plt.figure(num=1, figsize=(10,10))

plt.style.use('ggplot')

plt.plot(x, y, 'ro', x, y)
plt.xticks(x)
plt.yticks(y)
plt.show()
```
