```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
```

#Importamos los datos

```python
data = pd.read_csv('~/Documentos/Datasets/declaraciones-juradas-2014.csv')
```

```python
data.columns
```

```python
cargos = data[['apellido_nombre', 'cargo', 'ingresos_neto_gastos', 'actividad_principal_ambito', 'organismo', 'total_bienes_final']].groupby(['organismo', 'actividad_principal_ambito', 'cargo', 'apellido_nombre']).mean()
cargos = cargos.sort_values(by='total_bienes_final', ascending=False).head(20)
cargos['total_bienes_final'] = cargos['total_bienes_final'].map('${:,.0f}'.format)
cargos['ingresos_neto_gastos'] = cargos['ingresos_neto_gastos'].map('${:,.0f}'.format)
cargos
#cargos.plot(figsize=(20, 5))
```

```python
con_licencia = data[data.goza_de_licencia == 'SI'][['apellido_nombre',
                                                    'cargo',
                                                    'ingresos_neto_gastos',
                                                    'total_bienes_final',
                                                    'desde',
                                                    'fecha_inicio_licencia'
                                                    ]]

con_licencia['fecha_inicio_licencia'] = (pd.to_datetime(con_licencia['fecha_inicio_licencia'], format='%Y%m', errors='coerce')
                                         .dt
                                         .strftime('%Y-%m'))

con_licencia.sort_values(by='fecha_inicio_licencia', ascending=True, inplace=True)
con_licencia['total_bienes_final'] = con_licencia['total_bienes_final'].map('${:,.0f}'.format)
con_licencia['ingresos_neto_gastos'] = con_licencia['ingresos_neto_gastos'].map('${:,.0f}'.format)
con_licencia
```
