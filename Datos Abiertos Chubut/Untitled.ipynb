```python
import pandas as pd
import re
```

```python
data = pd.read_csv('../../Chubut Datos Abiertos/data/Registro de Daños y Asistencia.csv')
```

```python
data.info()
```

```python
# Categorías y descripciones en una columna
data['words'] = data['categorias'] + " " + data['Descripción de los daños o asistencia solicitada']

# Eliminar nulos
data.words.dropna(inplace=True)

# Unir columna en un único texto separado por espacios
text = ' '.join(data.words.str.lower())

# Eliminar carácteres innecesarios " ) ( ! / \n $
text = re.sub("[\n)(\"/!$]", '', text)

# Dividir en palabras
wordlist = re.split("[- ,.]", text)

# Guardar palabras en array
words = [ word for word in wordlist ]

# Eliminamos valores repetidos y ordenamos
sorted(set(words))
```
