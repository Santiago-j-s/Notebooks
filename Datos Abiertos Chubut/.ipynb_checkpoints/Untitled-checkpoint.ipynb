```python
import pandas as pd
```

```python
data = pd.read_csv('../../Chubut Datos Abiertos/data/Registro de Daños y Asistencia.csv')
```

```python
data.info()
```

```python
data['words'] = data['categorias'].str.lower() + " " + data['Descripción de los daños o asistencia solicitada'].str.lower()
wordlist = data['words'].str.split("[- ,.]").dropna()

words = []
for categoria in wordlist:
    for word in categoria:
        words.append(word.replace("[a\(]", ''))
        
set(words)
```
