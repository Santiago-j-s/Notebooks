```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

#Obtenemos la info

```python
#filepath = '/home/santiago/Escritorio/gsts.tsv'
#df = pd.read_csv(filepath, sep="\t", parse_dates=['Fecha'])
df = pd.read_clipboard(sep="\t", parse_dates=['Fecha'])
df['Gastos'] = (df['Gastos']
    .replace('\$', '', regex=True)
    .replace('\.', '', regex=True)
    .replace(',', '.', regex=True).astype(float))

df['Ingresos'] = (df['Ingresos']
    .replace('\$', '', regex=True)
    .replace('\.', '', regex=True)
    .replace(',', '.', regex=True).astype(float))

df.info()
```

#Agrupamos por fecha

```python
%matplotlib inline
by_date = df.groupby('Fecha')
```

##Graficamos

```python
sums = by_date.sum()
gastos_mes = sums[sums.index.month == 8]['Gastos']

graph = gastos_mes.plot(kind='bar')

xlabels = gastos_mes.index.strftime('%m-%d')
graph.xaxis.set_ticklabels(xlabels)

plt.show()
#gastos_mes = sums[sums.index.month == 8]['Gastos']
#graph = gastos_mes.plot(kind='bar')
#xlabels = gastos_mes.index.strftime('%m-%d')
#graph.xaxis.set_ticklabels(xlabels)
#plt.show()
```

##Fechas en que más dinero he gastado

```python
fechas_malas = pd.DataFrame(sums.Gastos.sort_values(ascending=False).head())
pd.DataFrame(fechas_malas.Gastos.values, index=fechas_malas.index.strftime('%m/%d'), columns=['Gastos'])
```

##Fechas con ingresos

```python
fechas_buenas = pd.DataFrame(sums.sort_values('Ingresos', ascending=False).head().dropna())
df_ingresos = df[df.Ingresos.notnull()]
merge = fechas_buenas.merge(df_ingresos, how='inner', left_index=True, right_on='Fecha')
merge
```

##Detalle de esos días

```python
df_gastos = df[df.Gastos.notnull()]
merge = pd.merge(df_gastos, fechas_malas, left_on='Fecha', right_index=True, how='inner')
merge = merge.sort_values('Gastos_y', ascending=False)
merge = merge.rename(columns={'Gastos_x': 'Gastos', 'Gastos_y': 'Sum'})
merge[['Tipo', 'Fecha', 'Descripción', 'Gastos', 'Sum']]
```
