```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import scipy.misc

plt.style.use('ggplot')
```

```python
x = np.linspace(-10, 10, 100)
y = x ** 2 - 4 * x + 2

plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.xticks(np.arange(-10, 11, 1))
plt.plot(x, np.zeros(x.size), 'k', lw=2)
plt.plot(0, y)
plt.show()
```
