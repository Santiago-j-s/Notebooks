```python
import numpy as np

import matplotlib.pyplot as plt

import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

%matplotlib inline
```

```python
def func(x, y):
    return 9 - x ** 2 - y ** 2

def func2(x, y):
    return np.sin(x) + np.cos(y)

x = y = np.linspace(-10, 10, 100)
x, y = np.meshgrid(x, y)

z = func(x, y)
z2 = func2(x, y)
```
