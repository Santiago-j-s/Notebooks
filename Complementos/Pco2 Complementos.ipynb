```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
def func(x):
    return np.multiply(x, np.sin(x)) - 1

pi = np.pi
a = 0
b = pi/2

x = np.linspace(-pi, pi, 1000)
y = func(x)

p0 = 0.7853
p1 = 1.1780
p2 = 0.9817
p3 = 1.0800

fig = plt.figure(figsize=(20, 6))
plt.plot(x, y, linewidth=2)

fig.suptitle('Método de la Bisección')

plt.axhline(0, color='k')
plt.plot((b, b), (0, func(b)), linewidth=2, color="r")

plt.plot((p0, p0), (0, func(p0)), color="k", linewidth=1)
plt.plot((p1, p1), (0, func(p1)), color="k", linewidth=1)
plt.plot((p2, p2), (0, func(p2)), color="k", linewidth=1)
plt.plot((p3, p3), (0, func(p3)), color="k", linewidth=1)

t_p0 = "{:.2f}".format(func(p0))
t_p1 = "{:.2f}".format(func(p1))
t_p2 = "{:.2f}".format(func(p2))
t_p3 = "{:.2f}".format(func(p3))

plt.text(p0-0.05, func(p0)-0.15, float(t_p0))
plt.text(p1-0.05, func(p1)+0.05, float(t_p1))
plt.text(p2-0.05, func(p2)-0.15, float(t_p2))
plt.text(p3-0.05, func(p3)-0.15, float(t_p3))

plt.xticks([0, p0, p1, p2, p3, b, pi], ["0", "p0", "p1", "p2", "p3", u"\u03C0/2", u"\u03C0"])

legend = " p0: {:.2f}\n p1: {:.2f}\n p2: {:.2f}\n p3: {:.2f}".format(p0, p1, p2, p3) 

plt.text(.05, .60, legend, fontsize=12)

plt.show()
```
