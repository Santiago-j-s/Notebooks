```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

colors = [color['color'] for color in list(plt.rcParams['axes.prop_cycle'])]

histconf = {
    'alpha': 0.9,
    'bins': 40,
    'color': colors[1],
}

def muestrear(muestra: np.ndarray, n: int, p: int) -> np.ndarray:
    """ Devuelve las medias de un conjunto de submuestras a partir de una muestra dada
        
        Args:
            muestra (ndarray): la muestra original
            n (int): cantidad de submuestras a generar
            p (int): cantidad de elementos por submuestra
        
        Returns:
            out (ndarray): las medias de cada submuestra
    """
    # n submuestras de p elementos a partir de muestra, con reposición
    samples = np.array([np.random.choice(muestra, p, replace=True) for i in range(n)])
    # media aritmética de cada submuestra
    means = np.array([np.mean(sample) for sample in samples])
    return means
```

# Teorema Central del Límite

Sea $ X_1, X_2, ..., X_n $ un conjunto de variables aleatorias independientes e idénticamente distribuidas de una distribución con media $ \mu $ y varianza $ \sigma $ entonces si $ n $ es suficientemente grande la variable aleatoria

$$ \overline{X} = \frac{1}{n} \sum\limits_{i=1}^{n} X_{i} $$

tiene una distribución normal con $ \mu_{\overline{X}} = \mu $ y $ \sigma^{2}_{\overline{X}} = \frac{\sigma}{n} $.

```python
def hist(x, subplot, title=None, xlabel=None, ylabel=None):
    plt.subplot(2, 3, subplot)
    if title is not None:
        plt.title(title)
    if ylabel is not None:
        plt.ylabel(ylabel)
    if xlabel is not None:
        plt.xlabel(xlabel)
        
    plt.hist(x, **histconf)
    
def graficar(dist, muestras):
    plt.figure(figsize=(14, 8))
    plt.subplots_adjust(hspace=0.5) # ajusta el espacio vertical entre subplots

    hist(dist, 1, title='Sin muestreo', ylabel='frecuencia')
    for i, muestra in enumerate(muestras, start=1):
        title = ''.join([str(len(muestra)), ' muestras'])
        ylabel = None if (i % 3 != 0) else 'frecuencia' # ylabel al inicio de cada fila
        xlabel = None if (i < 3) else 'valores'
        hist(muestra, i+1, title=title, ylabel=ylabel, xlabel=xlabel)
    
# Muestreo de 1000 elementos de una distribución aleatoria
r = np.random.random_sample(1000);

# Cantidad de elementos por submuestra
elems = 20

# Tamaño de cada submuestra a generar
N = [10, 100, 1000, 10000, 100000]

medias = [muestrear(r, n, elems) for n in N]
graficar(r, medias)

for i, media in enumerate(medias):
    print('media para {:d} muestras: {:{align}{width}.10f}'.format(N[i], np.mean(media), width=20-i, align='>'))
    print('desvío para {:d} muestras: {:{align}{width}.10f}'.format(N[i], np.sqrt(np.var(media)), width=19-i, align='>'))
    print()

plt.show()
```
