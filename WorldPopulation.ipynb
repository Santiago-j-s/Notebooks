```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
texts = []
with open('/home/santiago/Documentos/Datasets/worldbank/API_SP.POP.TOTL_DS2_en_csv_v2.csv') as f:
    for line in f:
        texts.append(line.split('","'))

for text in texts:
    text[0] = text[0][1::]
    del text[-1]
    
headers = texts[4]
data = texts[5::]
years = [int(year) for year in texts[4][4::]]
populations = [population[4::] for population in texts[5::]]
countries = [country[0] for country in texts[5::]]
```

```python
df = pd.DataFrame(data=data, columns=headers)
df
```

```python
fig = plt.figure(figsize=(15,5))
plt.plot(years, populations[257], label=countries[257])
plt.legend(loc=0)
plt.grid()
plt.show()
```
